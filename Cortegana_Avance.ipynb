```python
import requests
import re
import folium
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
from IPython.display import display

# Configuración de las APIs
GOOGLE_PLACES_API_KEY = "AIzaSyCTpUgmQvlXsxi7tAMUcqWusixHdo-YmcQ"
TMDB_API_KEY = "52affe755f0d4a073ee6e5d8102a6db6"

def get_places_in_district(district, category):
    """
    Consulta Google Places para obtener negocios de una categoría específica en un distrito.
    """
    endpoint = f"https://maps.googleapis.com/maps/api/place/textsearch/json"
    params = {
        "query": f"{category} in {district}",
        "key": GOOGLE_PLACES_API_KEY
    }
    response = requests.get(endpoint, params=params)
    if response.status_code == 200:
        results = response.json().get("results", [])
        return results
    return []

def filter_cinemas_with_regex(places, pattern):
    """
    Filtra los lugares usando expresiones regulares.
    """
    regex = re.compile(pattern, re.IGNORECASE)
    return [place for place in places if regex.search(place["name"])]

def filter_places_by_radius(places, center_coords, radius_km):
    """
    Filtra los lugares para que estén dentro de un radio especificado desde el centro.
    """
    filtered_places = []
    for place in places:
        place_coords = (
            place["geometry"]["location"]["lat"],
            place["geometry"]["location"]["lng"]
        )
        distance = geodesic(center_coords, place_coords).km
        if distance <= radius_km:
            filtered_places.append(place)
    return filtered_places

def get_tmdb_movies():
    """
    Consulta TMDB para obtener las películas en cartelera.
    """
    endpoint = f"https://api.themoviedb.org/3/movie/now_playing"
    params = {
        "api_key": TMDB_API_KEY,
        "language": "en-US",
        "region": "US"
    }
    response = requests.get(endpoint, params=params)
    if response.status_code == 200:
        results = response.json().get("results", [])
        return results
    return []

def display_map(places):
    """
    Crea un mapa interactivo con las ubicaciones de los lugares.
    """
    if not places:
        print("No hay lugares para mostrar en el mapa.")
        return

    # Coordenadas iniciales (primer lugar en la lista)
    first_place = places[0]
    lat = first_place["geometry"]["location"]["lat"]
    lng = first_place["geometry"]["location"]["lng"]

    # Crear el mapa
    map_obj = folium.Map(location=[lat, lng], zoom_start=13)

    # Agregar marcadores
    for place in places:
        name = place["name"]
        lat = place["geometry"]["location"]["lat"]
        lng = place["geometry"]["location"]["lng"]
        folium.Marker(location=[lat, lng], popup=name).add_to(map_obj)

    # Mostrar el mapa
    display(map_obj)

# Ejecución del flujo del programa
def main():
    # Paso 1: Definir los distritos
    tourist_districts = ["Barranco", "Miraflores", "San Borja", "Surco", "La Molina"]
    non_tourist_districts = ["El Agustino", "Santa Anita", "Ate", "Villa El Salvador", "Villa María del Triunfo", "San Juan de Lurigancho"]

    # Paso 2: Escoger el distrito
    print("Distritos turísticos disponibles:")
    print(", ".join(tourist_districts))
    print("Distritos no céntricos disponibles:")
    print(", ".join(non_tourist_districts))

    selected_district = input("Ingrese el distrito que desea explorar: ")

    # Obtener coordenadas del distrito
    geolocator = Nominatim(user_agent="geoapi")
    location = geolocator.geocode(selected_district)
    if not location:
        print("No se pudo encontrar el distrito seleccionado.")
        return

    center_coords = (location.latitude, location.longitude)
    radius_km = 5.0  # Definir un radio fijo de 5 kilómetros

    # Paso 3: Mostrar categorías y seleccionar una
    category = "cinema"
    print(f"Buscando lugares de la categoría: {category}")
    
    places = get_places_in_district(selected_district, category)

    # Paso 4: Filtrar los cines usando expresiones regulares
    filtered_places = filter_cinemas_with_regex(places, pattern="cine|cinema")

    # Paso 5: Filtrar los lugares por radio
    filtered_places = filter_places_by_radius(filtered_places, center_coords, radius_km)

    # Paso 6: Mostrar los cines en un mapa
    display_map(filtered_places)

    # Paso 7: Seleccionar un cine
    for i, place in enumerate(filtered_places):
        print(f"{i + 1}. {place['name']}")
    selected_index = int(input("Seleccione un cine por su número: ")) - 1
    selected_cinema = filtered_places[selected_index]

    # Paso 8: Mostrar cartelera de películas
    movies = get_tmdb_movies()
    print("Películas en cartelera:")
    for movie in movies:
        print(f"- {movie['title']} (Fecha de estreno: {movie['release_date']})")

if __name__ == "__main__":
    main()
```
