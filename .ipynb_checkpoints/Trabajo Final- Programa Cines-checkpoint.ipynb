```python
import requests
import re
import folium
from IPython.display import display

# Configuración de las APIs
GOOGLE_PLACES_API_KEY = "AIzaSyCTpUgmQvlXsxi7tAMUcqWusixHdo-YmcQ"
TMDB_API_KEY = "52affe755f0d4a073ee6e5d8102a6db6"

def get_places_in_district(district, category):
    """
    Consulta Google Places para obtener negocios de una categoría específica en un distrito.
    """
    endpoint = f"https://maps.googleapis.com/maps/api/place/textsearch/json"
    params = {
        "query": f"{category} in {district}",
        "key": GOOGLE_PLACES_API_KEY
    }
    response = requests.get(endpoint, params=params)
    if response.status_code == 200:
        results = response.json().get("results", [])
        return results
    return []

def filter_cinemas_with_regex(places, pattern):
    """
    Filtra los lugares usando expresiones regulares.
    """
    regex = re.compile(pattern, re.IGNORECASE)
    return [place for place in places if regex.search(place["name"])]

def get_tmdb_movies():
    """
    Consulta TMDB para obtener las películas en cartelera.
    """
    endpoint = f"https://api.themoviedb.org/3/movie/now_playing"
    params = {
        "api_key": TMDB_API_KEY,
        "language": "en-US",
        "region": "US"
    }
    response = requests.get(endpoint, params=params)
    if response.status_code == 200:
        results = response.json().get("results", [])
        return results
    return []

def display_map(places):
    """
    Crea un mapa interactivo con las ubicaciones de los lugares.
    """
    if not places:
        print("No hay lugares para mostrar en el mapa.")
        return

    # Coordenadas iniciales (primer lugar en la lista)
    first_place = places[0]
    lat = first_place["geometry"]["location"]["lat"]
    lng = first_place["geometry"]["location"]["lng"]

    # Crear el mapa
    map_obj = folium.Map(location=[lat, lng], zoom_start=13)

    # Agregar marcadores
    for place in places:
        name = place["name"]
        lat = place["geometry"]["location"]["lat"]
        lng = place["geometry"]["location"]["lng"]
        folium.Marker(location=[lat, lng], popup=name).add_to(map_obj)

    # Mostrar el mapa
    display(map_obj)

# Ejecución del flujo del programa
def main():
    # Paso 1: Definir los distritos
    tourist_districts = ["Barranco", "Miraflores", "San Borja", "Surco", "La Molina"]
    non_tourist_districts = ["El Agustino", "Santa Anita", "Ate", "Villa El Salvador", "Villa María del Triunfo", "San Juan de Lurigancho"]

    # Paso 2: Escoger el distrito
    print("Distritos turísticos disponibles:")
    print(", ".join(tourist_districts))
    print("Distritos no céntricos disponibles:")
    print(", ".join(non_tourist_districts))

    selected_district = input("Ingrese el distrito que desea explorar: ")

    # Paso 3: Mostrar categorías y seleccionar una
    category = "cinema"
    print(f"Buscando lugares de la categoría: {category}")
    
    places = get_places_in_district(selected_district, category)

    # Paso 4: Filtrar los cines usando expresiones regulares
    filtered_places = filter_cinemas_with_regex(places, pattern="cine|cinema")

    # Paso 5: Mostrar los cines en un mapa
    display_map(filtered_places)

    # Paso 6: Seleccionar un cine
    for i, place in enumerate(filtered_places):
        print(f"{i + 1}. {place['name']}")
    selected_index = int(input("Seleccione un cine por su número: ")) - 1
    selected_cinema = filtered_places[selected_index]

    # Paso 7: Mostrar cartelera de películas
    movies = get_tmdb_movies()
    print("Películas en cartelera:")
    for movie in movies:
        print(f"- {movie['title']} (Fecha de estreno: {movie['release_date']})")

if __name__ == "__main__":
    main()
```

```python
import requests
import re
import folium
import csv
from IPython.display import display

# Configuración de las APIs
GOOGLE_PLACES_API_KEY = "AIzaSyCTpUgmQvlXsxi7tAMUcqWusixHdo-YmcQ" 
# Es necesaria para autentificarnos con los servicios de Google
TMDB_API_KEY = "52affe755f0d4a073ee6e5d8102a6db6"

def get_places_in_district(district, category):
    """
    Consulta Google Places para obtener negocios de una categoría específica en un distrito.
    """
    endpoint = f"https://maps.googleapis.com/maps/api/place/textsearch/json"
    # Especifica la URL de la API de Google Places
    params = {
        "query": f"{category} in {district}",
        "key": GOOGLE_PLACES_API_KEY
    }
    # Define los parametros de la consulta que se enviaran como parte de la solicitud HTTP
    response = requests.get(endpoint, params=params)
    # Se configura la solicitud para que cuando se envie el endpoint, la API de Google procese la 
    # consulta especifica en params y devuelva luegares relevantesque coicidan con los criterios
    
    if response.status_code == 200:
        # Verifica si la solicitud HTTP a la API fue exitosa
        #   Cuadno se realiza la solicitud HTTP, la respuesta contiene un código de estado
        #   Un código de estado 200 significa que la solicitud fue procesada correctamente
        #    por el servidor
        #   Si el código no es 200, indica algún error, como problemas con la autenticación, formato
        #    incorrecto en los parámetros o errores del servidor
        
        results = response.json().get("results", [])
        # Convierte la respuesta HTTP (que esta en formato JSON) en un diccionario de Python
        #  '.get("results",[])'
        #    Busca la clave 'results'dentro del diccionario devuelto
        #    Si la clave 'results' existe, devuelve su valor
        #    Si no existe, devuelve una lista vacía
        
        return results
    return [] # Lista de resultados que incluye detalles de los negocios como nombre, dirrecion y coordenadas geograficas

def save_places_to_csv(places, filename):
    """
    Guarda los lugares obtenidos en un archivo CSV.
    """
    with open(filename, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(["Name", "Address", "Latitude", "Longitude"])
        for place in places:
            name = place.get("name", "")
            address = place.get("formatted_address", "")
            lat = place["geometry"]["location"].get("lat", "")
            lng = place["geometry"]["location"].get("lng", "")
            writer.writerow([name, address, lat, lng])

def filter_cinemas_with_regex(places, pattern):
    """
    Filtra los lugares usando expresiones regulares.
    """
    regex = re.compile(pattern, re.IGNORECASE)
    return [place for place in places if regex.search(place["name"])]

def get_tmdb_movies():
    """
    Consulta TMDB para obtener las películas en cartelera.
    """
    endpoint = f"https://api.themoviedb.org/3/movie/now_playing"
    params = {
        "api_key": TMDB_API_KEY,
        "language": "en-US",
        "region": "US"
    }
    response = requests.get(endpoint, params=params)
    if response.status_code == 200:
        results = response.json().get("results", [])
        return results
    return []

def save_movies_to_csv(movies, filename):
    """
    Guarda las películas obtenidas en un archivo CSV.
    """
    with open(filename, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(["Title", "Release Date", "Overview"])
        for movie in movies:
            title = movie.get("title", "")
            release_date = movie.get("release_date", "")
            overview = movie.get("overview", "")
            writer.writerow([title, release_date, overview])

def display_map(places):
    """
    Crea un mapa interactivo con las ubicaciones de los lugares.
    """
    if not places:
        print("No hay lugares para mostrar en el mapa.")
        return

    # Coordenadas iniciales (primer lugar en la lista)
    first_place = places[0]
    lat = first_place["geometry"]["location"].get("lat", 0)
    lng = first_place["geometry"]["location"].get("lng", 0)

    # Crear el mapa
    map_obj = folium.Map(location=[lat, lng], zoom_start=13)

    # Agregar marcadores
    for place in places:
        name = place.get("name", "")
        lat = place["geometry"]["location"].get("lat", 0)
        lng = place["geometry"]["location"].get("lng", 0)
        folium.Marker(location=[lat, lng], popup=name).add_to(map_obj)

    # Mostrar el mapa
    display(map_obj)

# Ejecución del flujo del programa
def main():
    # Paso 1: Definir los distritos
    tourist_districts = ["Barranco", "Miraflores", "San Borja", "Surco", "La Molina"]
    non_tourist_districts = ["El Agustino", "Santa Anita", "Ate", "Villa El Salvador", "Villa María del Triunfo", "San Juan de Lurigancho"]

    # Paso 2: Escoger el distrito
    print("Distritos turísticos disponibles:")
    print(", ".join(tourist_districts))
    print("Distritos no céntricos disponibles:")
    print(", ".join(non_tourist_districts))

    selected_district = input("Ingrese el distrito que desea explorar: ")

    # Paso 3: Mostrar categorías y seleccionar una
    category = "cinema"
    print(f"Buscando lugares de la categoría: {category}")
    
    places = get_places_in_district(selected_district, category)

    # Paso 4: Filtrar los cines usando expresiones regulares
    filtered_places = filter_cinemas_with_regex(places, pattern="cine|cinema")

    # Guardar lugares en un archivo CSV
    save_places_to_csv(filtered_places, "places.csv")

    # Paso 5: Mostrar los cines en un mapa
    display_map(filtered_places)

    # Paso 6: Seleccionar un cine
    for i, place in enumerate(filtered_places):
        print(f"{i + 1}. {place['name']}")
    selected_index = int(input("Seleccione un cine por su número: ")) - 1
    selected_cinema = filtered_places[selected_index]

    # Paso 7: Mostrar cartelera de películas
    movies = get_tmdb_movies()
    print("Películas en cartelera:")
    for movie in movies:
        print(f"- {movie['title']} (Fecha de estreno: {movie['release_date']})")

    # Guardar películas en un archivo CSV
    save_movies_to_csv(movies, "movies.csv")

if __name__ == "__main__":
    main()
```
